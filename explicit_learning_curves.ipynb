### Plot function

```python
def plot_learning_curve(score_dict):
    
    keys = score_dict["train"].keys()
    keys.sort()
    colors = sns.color_palette("husl", len(score_dict["train"].keys()))
    plt.figure(figsize=(10,6))
    
    for idx, key in enumerate(keys):
        plt.plot(score_dict["train_size"], score_dict["train"][key].mean(1), c=colors[idx],
                 label="%s %i" % (score_dict["variable"], key))
        plt.plot(score_dict["train_size"], score_dict["test"][key].mean(1), c=colors[idx])
    plt.xlabel("Training set size")
    plt.ylabel("MCC score")
    plt.ylim(0,1)
    plt.xlim(0,700000)
    plt.legend(loc="best")
```

## Validation set

```python
X_ref = np.load('../data/X_ref.npy')
y_ref = np.load('../data/y_ref.npy')
X_ref = np.hstack((X_ref[:,300:320], X_ref[:,330:339], X_ref[:,341:342]))
```

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import matthews_corrcoef
```

### Random Forest

```python
ref_scores = []
mcc = []
depths = range(1,31)
for depth in depths:
    rf_clf_depth = RandomForestClassifier(n_estimators=50, n_jobs=8, max_depth=depth,
                                          random_state=42, class_weight="auto")
    rf_clf_depth.fit(X_aa_ap_co, y_clstr)
    y_ref_pred = rf_clf_depth.predict(X_ref)
    ref_scores.append(precision_recall_fscore_support(y_ref, y_ref_pred, average='binary'))
    mcc.append(matthews_corrcoef(y_ref, y_ref_pred))
```

```python
precision = []
recall = []
f1 = []
for scores in ref_scores:
    precision.append(scores[0])
    recall.append(scores[1])
    f1.append(scores[2])
```

```python
plt.plot(depths, precision, label="Precision")
plt.plot(depths, recall, label="Recall")
plt.plot(depths, f1, label="F1")
plt.plot(depths, mcc, label="MCC")
plt.legend(loc='best')
plt.savefig('random_forest_scores_vary_depth.pdf', bbox_inches='tight')
plt.show()
```

### Extra Random Trees

```python
et_ref_scores = []
et_mcc = []
depths = range(1,31)
for depth in depths:
    et_clf_depth = ExtraTreesClassifier(n_estimators=50, n_jobs=8, max_depth=depth,
                                        random_state=42, class_weight="auto")
    et_clf_depth.fit(X_aa_ap_co, y_clstr)
    y_ref_pred = et_clf_depth.predict(X_ref)
    et_ref_scores.append(precision_recall_fscore_support(y_ref, y_ref_pred, average='binary'))
    et_mcc.append(matthews_corrcoef(y_ref, y_ref_pred))
```

```python
et_precision = []
et_recall = []
et_f1 = []
for scores in et_ref_scores:
    et_precision.append(scores[0])
    et_recall.append(scores[1])
    et_f1.append(scores[2])
```

```python
plt.plot(depths, et_precision, label="Precision")
plt.plot(depths, et_recall, label="Recall")
plt.plot(depths, et_f1, label="F1")
plt.plot(depths, et_mcc, label="MCC")
plt.legend(loc='best')
plt.savefig('extra_random_trees_scores_vary_depth.pdf', bbox_inches='tight')
plt.show()
```

## Cross validation score for different depths

```python
cv_scores = {}
depths = [5, 10, 15, 20, 25, 30]
for depth in depths:
    cv_scores[depth] = {'precision': [], 'recall': [], 'fscore': [], 'mcc': []}
    
    rf_clf_depth = RandomForestClassifier(n_estimators=50, n_jobs=8, max_depth=depth,
                                          random_state=42, class_weight="auto")
    for train, test in cv:
        rf_clf_depth.fit(X_aa_ap_co[train], y_clstr[train])
        y_clstr_pred = rf_clf_depth.predict(X_aa_ap_co[test])
        
        precision, recall, fscore, _ = precision_recall_fscore_support(y_clstr[test], y_clstr_pred, average='binary')
        mcc_score = matthews_corrcoef(y_clstr[test], y_clstr_pred)
        
        cv_scores[depth]['precision'].append(precision)
        cv_scores[depth]['recall'].append(recall)
        cv_scores[depth]['fscore'].append(fscore)
        cv_scores[depth]['mcc'].append(mcc_score)
```

```python
plt.plot(depths, [np.array(cv_scores[depth]['precision']).mean() for depth in depths], label='Precision')
plt.plot(depths, [np.array(cv_scores[depth]['recall']).mean() for depth in depths], label='Recall')
plt.plot(depths, [np.array(cv_scores[depth]['fscore']).mean() for depth in depths], label='F1')
plt.plot(depths, [np.array(cv_scores[depth]['mcc']).mean() for depth in depths], label='MCC')
plt.legend(loc='best')
```
