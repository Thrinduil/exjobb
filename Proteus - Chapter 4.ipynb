```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set(context='paper', style='darkgrid', palette='deep', font='sans-serif', font_scale=1.2, rc=None)
```

```python
import pickle
def save_obj(obj, name):
    with open("../data/" + name + ".pkl", "wb") as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
def load_obj(name):
    with open("../data/" + name + ".pkl", "rb") as f:
        return pickle.load(f)
```

# 4.1 Amino Acid Concentration

## For protean/non-protean regions

```python
import itertools

records = load_obj("records")
all_seqs = []
pros_seqs = []
morf_seqs = []
all_seqs_protean = []
pros_seqs_protean = []
morf_seqs_protean = []
all_seqs_nonprotean = []
pros_seqs_nonprotean = []
morf_seqs_nonprotean = []

for name in records.keys():
    all_seqs.extend(records[name]["sequence"])
    all_seqs_protean.extend(list(itertools.compress(records[name]["sequence"], [int(label) for label in records[name]["labels"]])))
    all_seqs_nonprotean.extend(list(itertools.compress(records[name]["sequence"], [(int(label)+1)%2 for label in records[name]["labels"]])))
    
    if name[0] == "I":
        pros_seqs.extend(records[name]["sequence"])
        pros_seqs_protean.extend(list(itertools.compress(records[name]["sequence"], [int(label) for label in records[name]["labels"]])))
        pros_seqs_nonprotean.extend(list(itertools.compress(records[name]["sequence"], [(int(label)+1)%2 for label in records[name]["labels"]])))
        
    elif name[0] == "M":
        morf_seqs.extend(records[name]["sequence"])
        morf_seqs_protean.extend(list(itertools.compress(records[name]["sequence"], [int(label) for label in records[name]["labels"]])))
        morf_seqs_nonprotean.extend(list(itertools.compress(records[name]["sequence"], [(int(label)+1)%2 for label in records[name]["labels"]])))
```

```python
aa_list = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
           'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
all_percentages = []
morf_percentages = []
pros_percentages = []
all_protean_percentages = []
morf_protean_percentages = []
pros_protean_percentages = []
all_nonprotean_percentages = []
morf_nonprotean_percentages = []
pros_nonprotean_percentages = []

for aa in aa_list:
    all_percentages.append(float(all_seqs.count(aa)) / len(all_seqs))
    morf_percentages.append(float(morf_seqs.count(aa)) / len(morf_seqs))
    pros_percentages.append(float(pros_seqs.count(aa)) / len(pros_seqs))
    
    all_protean_percentages.append(float(all_seqs_protean.count(aa)) / len(all_seqs_protean))
    morf_protean_percentages.append(float(morf_seqs_protean.count(aa)) / len(morf_seqs_protean))
    pros_protean_percentages.append(float(pros_seqs_protean.count(aa)) / len(pros_seqs_protean))
    
    all_nonprotean_percentages.append(float(all_seqs_nonprotean.count(aa)) / len(all_seqs_nonprotean))
    morf_nonprotean_percentages.append(float(morf_seqs_nonprotean.count(aa)) / len(morf_seqs_nonprotean))
    pros_nonprotean_percentages.append(float(pros_seqs_nonprotean.count(aa)) / len(pros_seqs_nonprotean))
```

```python
all_diff = np.array(all_nonprotean_percentages) - np.array(all_protean_percentages)
morf_diff = np.array(morf_nonprotean_percentages) - np.array(morf_protean_percentages)
pros_diff = np.array(pros_nonprotean_percentages) - np.array(pros_protean_percentages)
```

```python
data_protean_diff = {"Amino acid": aa_list * 3,
                     "Difference": np.ravel(np.vstack((all_diff, morf_diff, pros_diff))),
                     "Dataset": (["MoRF & ProS"] * 20 + ["MoRF"] * 20 + ["ProS"] * 20)}
data_protean_diff = pd.DataFrame(data_protean_diff)
```

```python
sns.factorplot("Amino acid", "Difference", "Dataset", data=data_protean_diff, kind="bar", size=4.5, aspect=7./4.5, legend_out=False,
               hue_order=["MoRF", "ProS", "MoRF & ProS"],
               x_order=['A', 'V', 'I', 'L', 'M', 'F', 'Y', 'W', 'S', 'T', 'N', 'Q', 'R', 'H', 'K', 'D', 'E', 'C', 'G', 'P'])
plt.savefig('amino_acid_protean_diff.pdf', bbox_inches='tight')
```

## For order/disorder regions

```python
import itertools
all_seqs = []
pros_seqs = []
morf_seqs = []
all_seqs_disorder = []
pros_seqs_disorder = []
morf_seqs_disorder = []
all_seqs_order = []
pros_seqs_order = []
morf_seqs_order = []
for name in records.keys():
    
    filename = "../disopred/output3/" + name + ".diso"
    disorder_cutoff = 0.5
    
    with open(filename) as f:
        lines = f.readlines()[3:]
    disorder = []
    for line in lines:
        score = float(line.split()[3])
        if score >= disorder_cutoff:
            disorder.append(1)
        if score < disorder_cutoff:
            disorder.append(0)
        
    all_seqs.extend(records[name]["sequence"])
    all_seqs_disorder.extend(list(itertools.compress(records[name]["sequence"], disorder)))
    all_seqs_order.extend(list(itertools.compress(records[name]["sequence"], [(label+1)%2 for label in disorder])))
    if name[0] == "I":
        pros_seqs.extend(records[name]["sequence"])
        pros_seqs_disorder.extend(list(itertools.compress(records[name]["sequence"], disorder)))
        pros_seqs_order.extend(list(itertools.compress(records[name]["sequence"], [(label+1)%2 for label in disorder])))

    elif name[0] == "M":
        morf_seqs.extend(records[name]["sequence"])
        morf_seqs_disorder.extend(list(itertools.compress(records[name]["sequence"], disorder)))
        morf_seqs_order.extend(list(itertools.compress(records[name]["sequence"], [(label+1)%2 for label in disorder])))
```

```python
aa_list = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
           'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
all_percentages = []
morf_percentages = []
pros_percentages = []
all_disorder_percentages = []
morf_disorder_percentages = []
pros_disorder_percentages = []
all_order_percentages = []
morf_order_percentages = []
pros_order_percentages = []

for aa in aa_list:
    all_percentages.append(float(all_seqs.count(aa)) / len(all_seqs))
    morf_percentages.append(float(morf_seqs.count(aa)) / len(morf_seqs))
    pros_percentages.append(float(pros_seqs.count(aa)) / len(pros_seqs))
    
    all_disorder_percentages.append(float(all_seqs_disorder.count(aa)) / len(all_seqs_disorder))
    morf_disorder_percentages.append(float(morf_seqs_disorder.count(aa)) / len(morf_seqs_disorder))
    pros_disorder_percentages.append(float(pros_seqs_disorder.count(aa)) / len(pros_seqs_disorder))
    
    all_order_percentages.append(float(all_seqs_order.count(aa)) / len(all_seqs_order))
    morf_order_percentages.append(float(morf_seqs_order.count(aa)) / len(morf_seqs_order))
    pros_order_percentages.append(float(pros_seqs_order.count(aa)) / len(pros_seqs_order))
```

```python
all_disorder_diff = np.array(all_order_percentages) - np.array(all_disorder_percentages)
morf_disorder_diff = np.array(morf_order_percentages) - np.array(morf_disorder_percentages)
pros_disorder_diff = np.array(pros_order_percentages) - np.array(pros_disorder_percentages)
```

```python
data_disorder_diff = {"Amino acid": aa_list * 3,
                      "Difference": np.ravel(np.vstack((all_disorder_diff, morf_disorder_diff, pros_disorder_diff))),
                      "Dataset": (["MoRF & ProS"] * 20 + ["MoRF"] * 20 + ["ProS"] * 20)}
data_disorder_diff = pd.DataFrame(data_disorder_diff)
%store data_disorder_diff
```

```python
sns.factorplot("Amino acid", "Difference", "Dataset", data=data_disorder_diff, kind="bar", size=4.5, aspect=7./4.5, legend_out=False,
               hue_order=["MoRF", "ProS", "MoRF & ProS"],
               x_order=['A', 'V', 'I', 'L', 'M', 'F', 'Y', 'W', 'S', 'T', 'N', 'Q', 'R', 'H', 'K', 'D', 'E', 'C', 'G', 'P'])
plt.savefig('amino_acid_disorder_diff.pdf', bbox_inches='tight')
```

# 4.3 Conservation

```python
def get_conservation_features(name):

    filename = "../disopred/output3/" + name + ".psi"
    with open(filename) as f:
        lines = f.readlines()[3:-6]
    con_scores = []
    for line in lines:
        words = line.split()
        con_scores.append([float(words[42].strip())])
    
    con_scores = np.array(con_scores)

    return con_scores
```

```python
import itertools
all_con = []
pros_con = []
morf_con = []
all_con_protean = []
pros_con_protean = []
morf_con_protean = []
all_con_nonprotean = []
pros_con_nonprotean = []
morf_con_nonprotean = []
for name in records.keys():
    con_scores = get_conservation_features(name).ravel()
    
    all_con.extend(con_scores)
    all_con_protean.extend(list(itertools.compress(con_scores, [int(label) for label in records[name]["labels"]])))
    all_con_nonprotean.extend(list(itertools.compress(con_scores, [(int(label)+1)%2 for label in records[name]["labels"]])))
    
    if name[0] == "I":
        pros_con.extend(con_scores)
        pros_con_protean.extend(list(itertools.compress(con_scores, [int(label) for label in records[name]["labels"]])))
        pros_con_nonprotean.extend(list(itertools.compress(con_scores, [(int(label)+1)%2 for label in records[name]["labels"]])))

    elif name[0] == "M":
        morf_con.extend(con_scores)
        morf_con_protean.extend(list(itertools.compress(con_scores, [int(label) for label in records[name]["labels"]])))
        morf_con_nonprotean.extend(list(itertools.compress(con_scores, [(int(label)+1)%2 for label in records[name]["labels"]])))
```

```python
data_con = {"Average conservation score": (all_con_nonprotean +
                                           all_con_protean +
                                           morf_con_nonprotean +
                                           morf_con_protean +
                                           pros_con_nonprotean +
                                           pros_con_protean),
             "Region": (["Non-protean"] * len(all_con_nonprotean) + ["Protean"] * len(all_con_protean) +
                        ["Non-protean"] * len(morf_con_nonprotean) + ["Protean"] * len(morf_con_protean) +
                        ["Non-protean"] * len(pros_con_nonprotean) + ["Protean"] * len(pros_con_protean)),
             "Dataset": (["MoRF & ProS"] * len(all_con) + ["MoRF"] * len(morf_con) + ["ProS"] * len(pros_con))}
data_con = pd.DataFrame(data_con)
%store data_con
```

```python
g = sns.factorplot("Dataset", "Average conservation score", "Region", data=data_con,
                   kind="bar", size=3.8, aspect=7./3.8, legend_out=False, n_boot=1000,
                   x_order=["MoRF", "ProS", "MoRF & ProS"],
                   hue_order=["Non-protean", "Protean"])
g.set(ylim=(0, 1))
plt.savefig('average_conservation.pdf', bbox_inches='tight')
```

# 4.4 Disorder Prediction

```python
fold_names = load_obj("fold_names")
X = np.load('../data/X_ws_15.npy')
current_idx = 0
pros_morf_both = []
pros_morf_protean = []
pros_morf_nonprotean = []
pros_both = []
pros_protean = []
pros_nonprotean = []
morf_both = []
morf_protean = []
morf_nonprotean = []
for fold in range(len(fold_names)):
    for name in fold_names[fold]:
        disorder_scores = X[current_idx:current_idx + len(records[name]["labels"]), 320]
        current_idx += len(records[name]["labels"])
        
        pros_morf_both.extend(disorder_scores)
        pros_morf_protean.extend(disorder_scores[np.array([int(label) for label in records[name]["labels"]]).astype("bool")])
        pros_morf_nonprotean.extend(disorder_scores[np.logical_not(np.array([int(label) for label in records[name]["labels"]]).astype("bool"))])
        
        if name[:3] == "IID":
            pros_both.extend(disorder_scores)
            pros_protean.extend(disorder_scores[np.array([int(label) for label in records[name]["labels"]]).astype("bool")])
            pros_nonprotean.extend(disorder_scores[np.logical_not(np.array([int(label) for label in records[name]["labels"]]).astype("bool"))])
            
        if name[:3] == "MoR":
            morf_both.extend(disorder_scores)
            morf_protean.extend(disorder_scores[np.array([int(label) for label in records[name]["labels"]]).astype("bool")])
            morf_nonprotean.extend(disorder_scores[np.logical_not(np.array([int(label) for label in records[name]["labels"]]).astype("bool"))])
```

```python
disorder_data = {"Average disorder score": (morf_protean + morf_nonprotean +
                                            pros_protean + pros_nonprotean +
                                            pros_morf_protean + pros_morf_nonprotean),
                 "Region": (["Protean"] * len(morf_protean) + 
                            ["Non-protean"] * len(morf_nonprotean) +
                            ["Protean"] * len(pros_protean) + 
                            ["Non-protean"] * len(pros_nonprotean) +
                            ["Protean"] * len(pros_morf_protean) + 
                            ["Non-protean"] * len(pros_morf_nonprotean)),
                 "Dataset": (["MoRF"] * (len(morf_both)) +
                             ["ProS"] * (len(pros_both)) +
                             ["MoRF & ProS"] * (len(pros_morf_both)))}
disorder_data = pd.DataFrame(disorder_data)
```

```python
g = sns.factorplot("Dataset", "Average disorder score", "Region",
                   data=disorder_data, kind="bar", size=3.5, aspect=7./3.5, legend_out=False,
                   x_order=["MoRF", "ProS", "MoRF & ProS"],
                   hue_order=["Non-protean", "Protean"])
g.set(ylim=(0,0.7))
plt.savefig('average_disorder.pdf', bbox_inches='tight')
```

# 4.4.1 Location of Protean Regions

```python
def viz_protean_regions(name):
        
    with open("../disopred/output3/" + name + ".diso") as f:
        lines = f.readlines()[3:]
    scores = []
    for line in lines:
        words = line.split()
        scores.append(float(words[3].strip()))
        
    labels = np.array([int(label) for label in records[name]["labels"]])
        
    plt.figure(figsize=(7,3))
    plt.plot(range(len(scores)), scores, label=name)
    plt.legend(loc="best")
    plt.plot(range(len(scores)), [0.5] * len(scores), "r--")
    plt.fill_between(range(len(scores)),
                     labels * 1,
                     labels * 0,
                     color="g", alpha=.3)
    plt.ylim((0,1))
    plt.xlabel("Residue")
    plt.ylabel("Disorder prediction score")
    plt.savefig('dis_vis_' + name + '.pdf', bbox_inches='tight')
    plt.show()
```
