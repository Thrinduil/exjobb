```python
from json import dump
from xml.dom import minidom

# Parse the xml database to memory
xmldoc = minidom.parse(inFile)

# Get root element and then all protein entries
IDEAL = xmldoc.getElementsByTagName('IDEAL')[0]
Entries = IDEAL.getElementsByTagName('IDEAL_entry')

# Create dictionary for future use
ProSList = {}

# Loop through every protein entry and save information to the dictionary
for Entry in Entries:

    # Get regions child elements
    Regions = Entry.getElementsByTagName("Region")

    # Get protein name and create dictionary entry with that name
    name = str(Entry.getElementsByTagName("idp_id")[0].firstChild.data)
    ProSList[name] = {}

    # Create lists for later use
    ProSList[name]["orderRegions"] = []
    ProSList[name]["disorderRegions"] = []
    ProSList[name]["ProSFlag"] = []
    ProSList[name]["ProSRegions"] = []

    # Get sequence and put in dictionary
    seq = str(Entry.getElementsByTagName("sequence")[0].firstChild.data)
    ProSList[name]["sequence"] = seq

    # Get order/disorder regions
    for Region in Regions:
        regionFlag = str(Region.getElementsByTagName("order_disorder")[0] \
                                                        .firstChild.data)
        regionStart = int(Region.getElementsByTagName("region_start")[0] \
                                                        .firstChild.data)
        regionEnd = int(Region.getElementsByTagName("region_end")[0] \
                                                        .firstChild.data)

        # Put in order or disorder list, depending on order flag
        if regionFlag == "order":
            ProSList[name]["orderRegions"].append((regionStart,
                                                   regionEnd))
        elif regionFlag == "disorder":
            ProSList[name]["disorderRegions"].append((regionStart,
                                                      regionEnd))

    # Check if current protein contains ProS regions and save to hasPros
    if len(Entry.getElementsByTagName("function_type")) > 0:
        if Entry.getElementsByTagName("function_type")[0] \
                                    .firstChild.data == "ProS":
            ProSList[name]["hasProS"] = True
    else:
        ProSList[name]["hasProS"] = False

    ##### This section is only for proteins containing ProS regions #####

    if ProSList[name]["hasProS"] == True:

        # Get Functions child elements
        Functions = Entry.getElementsByTagName("Function")

        # Loop through all functions and get ProS regions and flag
        for Function in Functions:

            # Make sure it's a ProS function
            if Function.getElementsByTagName(
                            "function_type")[0].firstChild.data == "ProS":

                # Fill ProSFlag
                ProSFlag = str(Function.getElementsByTagName(
                                    "pros_type")[0].firstChild.data)
                ProSList[name]["ProSFlag"].append(ProSFlag)

                # Fill ProSRegions WARNING: ONLY ORDER!!!
                # Try statement because entries might be empty.
                try:
                    ProSStart = int(Function.getElementsByTagName(
                                "order_region_start")[0].firstChild.data)
                    ProSEnd = int(Function.getElementsByTagName(
                                "order_region_end")[0].firstChild.data)
                    ProSList[name]["ProSRegions"].append((ProSStart,
                                                          ProSEnd))
                except AttributeError:
                    pass

    ##### Code to read disorder regions as well. If need be. #####
#                    try:
#                        ProSStart = int(Function.getElementsByTagName(
#                                 "disorder_region_start")[0].firstChild.data)
#                        ProSEnd = int(Function.getElementsByTagName(
#                                 "disorder_region_end")[0].firstChild.data)
#                        ProSList[name]["ProSRegions"].append((ProSStart,
#                                                              ProSEnd))
#                    except AttributeError:
#                        pass
    ##### End of section only for proteins containing ProS regions. #####

with open(outFile, "w") as exportFile:
    dump(ProSList, exportFile)

return None
```
