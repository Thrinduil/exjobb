```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

# The Fantastic Features

Now that we know which features work best, let's dig a bit deeper into them.

```python
y_morf = np.load("../data/y_morf.npy")
X_myfeat_morf = np.load("../data/X_myfeat_morf.npy")
X_myfeat_morf_aa = X_myfeat_morf[:,300:320]
X_myfeat_morf_do = X_myfeat_morf[:,320:324]
X_myfeat_morf_aa_do = np.hstack((X_myfeat_morf_aa, X_myfeat_morf_do))

y_pros = np.load("../data/y_pros.npy")
X_myfeat_pros = np.load("../data/X_myfeat_pros.npy")
X_myfeat_pros_aa = X_myfeat_pros[:,300:320]
X_myfeat_pros_do = X_myfeat_pros[:,320:324]
X_myfeat_pros_aa_do = np.hstack((X_myfeat_pros_aa, X_myfeat_pros_do))

y_both = np.load("../data/y_both.npy")
X_myfeat_both = np.load("../data/X_myfeat_both.npy")
X_myfeat_both_aa = X_myfeat_both[:,300:320]
X_myfeat_both_do = X_myfeat_both[:,320:324]
X_myfeat_both_aa_do = np.hstack((X_myfeat_both_aa, X_myfeat_both_do))
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)

from sklearn.cross_validation import StratifiedKFold
cv_morf = StratifiedKFold(y_morf, n_folds=5,
                          shuffle=True, random_state=42)
cv_pros = StratifiedKFold(y_pros, n_folds=5,
                          shuffle=True, random_state=42)
cv_both = StratifiedKFold(y_both, n_folds=5,
                          shuffle=True, random_state=42)

from sklearn.cross_validation import cross_val_score

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
et_clf_scaler = make_pipeline(StandardScaler(),
                              ExtraTreesClassifier(n_estimators=50,
                                                   class_weight="auto",
                                                   random_state=42))
et_clf = ExtraTreesClassifier(n_estimators=50,
                              class_weight="auto",
                              random_state=42)
rf_clf = RandomForestClassifier(n_estimators=50,
                                class_weight="auto",
                                random_state=42)
et_clf_5jobs = ExtraTreesClassifier(n_estimators=50,
                                    class_weight="auto",
                                    random_state=42, n_jobs=5)
```

## The Scaler

Up to this point, we have scaled the features before feeding them into the classifier. But as we have settled on a descision tree classifier, that is no longer neccesary, as that type of classifier is completely unsensitive to whether the features are scaled or not. 

But let's run a test to be sure.

```python
%%time
et_noscale_morf_scores = cross_val_score(et_clf, X_myfeat_morf_aa_do, y_morf,
                                 scoring=mcc, cv=cv_morf, n_jobs=5)
et_scale_morf_scores = cross_val_score(et_clf_scaler, X_myfeat_morf_aa_do, y_morf,
                                 scoring=mcc, cv=cv_morf, n_jobs=5)
%store et_noscale_morf_scores
%store et_scale_morf_scores
```

```python
myfeat_data = {"mcc_score": np.hstack((et_noscale_morf_scores,
                                       et_scale_morf_scores)),
               "scaled": (["No"] * 5 +
                          ["Yes"] * 5)
              }
myfeat_data = pd.DataFrame(myfeat_data)
sns.factorplot("scaled", "mcc_score",
               data=myfeat_data, kind="bar")
```

As we can see, no difference at all. From now on, we will go on without the scaler.

## How Many Trees Make a Forest?

Another thng we should get out of the way is how many descision trees we should have in our forest. So far we have used 50 trees. In general, the more trees the better, but you get sharp diminishing returns after a critical number of trees.

```python
estimators = np.logspace(np.log2(1), np.log2(100), 10, base=2)
```

```python
%%time
trees_score = []
for n_estimators in estimators.astype(int):
    et_clf_trees = ExtraTreesClassifier(n_estimators=n_estimators,
                                        class_weight="auto",
                                        random_state=42)
    trees_score.append(cross_val_score(et_clf_trees, X_myfeat_morf_aa_do, y_morf,
                                       scoring=mcc, cv=cv_morf, n_jobs=5))
% store trees_score
```

```python
plt.errorbar(estimators.astype(int), np.array(trees_score).mean(axis=1), yerr=np.array(trees_score).std(axis=1))
plt.xscale("log")
plt.show()
```

50 trees seems to be a reasonable amount. We could probably get away with using fewer, but it is better to err on the safe side. (I don't know why the score suddenly dips at around 12 trees. Must be a fluke.)

## Learning Curves

Let's look at some learning curves for the three datasets. A learning curve is generated by feeding the classifier an increasing share of the whole dataset. For instance: first we feed it 10% of the whole dataset, then 50% and then 100%. We look at the both the test score (as usual), and also the training score, which is the score we get if we would ask the classifier to classify the data it had just trainded itself on.

The test score should increase the more data we feed it. The training score should decrease slightly as we increase the amount of data.

```python
def plot_learning_curve(train_sizes, train_scores, test_scores, title):

    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("MCC score")
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="lower right")
    return plt
```

```python
%%time
from sklearn.learning_curve import learning_curve
et_morf_train_sizes, et_morf_train_scores, et_morf_test_scores = learning_curve(
    et_clf, X_myfeat_morf_aa_do, y_morf, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_morf, scoring=mcc, n_jobs=5)
from sklearn.learning_curve import learning_curve
et_pros_train_sizes, et_pros_train_scores, et_pros_test_scores = learning_curve(
    et_clf, X_myfeat_pros_aa_do, y_pros, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_pros, scoring=mcc, n_jobs=5)
from sklearn.learning_curve import learning_curve
et_both_train_sizes, et_both_train_scores, et_both_test_scores = learning_curve(
    et_clf, X_myfeat_both_aa_do, y_both, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_both, scoring=mcc, n_jobs=5)
%store et_morf_train_sizes
%store et_morf_train_scores
%store et_morf_test_scores
%store et_pros_train_sizes
%store et_pros_train_scores
%store et_pros_test_scores
%store et_both_train_sizes
%store et_both_train_scores
%store et_both_test_scores
```

```python
morf_title = "Learning Curves (MoRF dataset, Extra Trees classifier)"
morf_learning_curve = plot_learning_curve(et_morf_train_sizes, et_morf_train_scores,
                    et_morf_test_scores, morf_title)
pros_title = "Learning Curves (ProS dataset, Extra Trees classifier)"
pros_learning_curve = plot_learning_curve(et_pros_train_sizes, et_pros_train_scores,
                    et_pros_test_scores, pros_title)
both_title = "Learning Curves (Both datasets, Extra Trees classifier)"
both_learning_curve = plot_learning_curve(et_both_train_sizes, et_both_train_scores,
                    et_both_test_scores, both_title)
```

None of the learning curves has flatten out before the last datapoint. That means we could probably get a better result if we had more examples to work with. But it is good to see that the cross-validation score is steadily improving.

## Feature Importance

```python
%%time
et_clf_5jobs.fit(X_myfeat_morf_aa_do, y_morf)
importances = et_clf_5jobs.feature_importances_
importances_array = np.zeros((1,1))
for tree in et_clf_5jobs.estimators_:
    importances_vector = np.array(tree.feature_importances_).reshape((24,1))
    importances_array = np.vstack((importances_array, importances_vector))
```

```python
feature_names = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
                 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y',
                 "score", "length", "start", "stop"]    
importances_data = {"importance": list(importances_array[1:,:]),
                    "feature": feature_names * 50}
importances_data = pd.DataFrame(importances_data)
indices = np.argsort(importances)[::-1]
sorted_feature_names = [feature_names[idx] for idx in indices]
sns.factorplot("feature", "importance", data=importances_data, kind="bar",
               aspect=2.5, x_order=sorted_feature_names)
```

The disorder prediction score seems to be the single most important feature in predicting. The rest of the disorder features are grouped at the back together with some rare amino acids.

```python
# Utility function to report best scores
from operator import itemgetter
def report(grid_scores, n_top=3):
    top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print("Model with rank: {0}".format(i + 1))
        print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
              score.mean_validation_score,
              np.std(score.cv_validation_scores)))
        print("Parameters: {0}".format(score.parameters))
        print("")
```
