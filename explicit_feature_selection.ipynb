```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
y_clstr = np.load("../data/y_clstr.npy")
X_clstr = np.load("../data/X_clstr.npy")
X_clstr_ps = X_clstr[:,:300]
X_clstr_aa = X_clstr[:,300:320]
X_clstr_do = X_clstr[:,320:324]
X_clstr_ss = X_clstr[:,324:327]
X_clstr_ex = X_clstr[:,327:336]
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)

from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
cv_clstr = KFold(len(y_clstr), n_folds=5, shuffle=False, random_state=42)

from sklearn.ensemble import ExtraTreesClassifier
et_clf = ExtraTreesClassifier(n_estimators=50, n_jobs=7, random_state=42, class_weight="auto")
```

```python
X_clstr_dict = {"ps": X_clstr_ps,
                "aa": X_clstr_aa,
                "do": X_clstr_do,
                "ss": X_clstr_ss,
                "ex": X_clstr_ex}
```

```python
%%time
from itertools import chain, combinations
def all_subsets(ss):
    return chain(*map(lambda x: combinations(ss, x), range(1, len(ss)+1)))

feature_tags = ["ps", "aa", "do", "ss", "ex"]
tags = []

clstr_features_scores = {}
for subset in all_subsets(feature_tags):
    tag = "_".join(list(subset))
    tags.append(tag)
    
    clstr_features = np.zeros((X_clstr_dict["ps"].shape[0], 1))
    for feat_set in subset:
        clstr_features = np.hstack((clstr_features, X_clstr_dict[feat_set]))  
    clstr_features = clstr_features[:,1:]
    
    clstr_features_scores[tag] = cross_val_score(et_clf, clstr_features, y_clstr,
                                                 scoring=mcc, cv=cv_clstr)

%store clstr_features_scores
%store tags
```

```python
feature_data = {"mcc_score": np.zeros((1,)),
                "dataset": [],
                "features": []}

for tag in tags:
    feature_data["mcc_score"] = np.hstack((feature_data["mcc_score"], clstr_features_scores[tag]))
    feature_data["dataset"] = feature_data["dataset"] + ["Cluster"] * 5
    feature_data["features"] = feature_data["features"] + [tag] * 5
    
feature_data["mcc_score"] = feature_data["mcc_score"][1:]

feature_data = pd.DataFrame(feature_data)
%store feature_data
```

```python
sns.factorplot("dataset", "mcc_score", "features",
               data=feature_data, kind="bar", aspect=2,
               palette="deep",
               hue_order=[features for features in (feature_data.groupby(["features"])
                                                      ["mcc_score"].mean().reset_index()
                                                      .sort("mcc_score", ascending=False)
                                                      ["features"])])
```
