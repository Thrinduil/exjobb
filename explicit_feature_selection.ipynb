```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set(context='paper', style='darkgrid', palette='deep', font='sans-serif', font_scale=1.2, rc=None)
```

```python
y_clstr = np.load("../data/y_clstr.npy")
X_clstr = np.load("../data/X_clstr.npy")
splits = np.load("../data/splits.npy")
X_clstr_ps = X_clstr[:,:300]
X_clstr_aa = X_clstr[:,300:320]
X_clstr_do = X_clstr[:,320:327]
X_clstr_ss = X_clstr[:,327:330]
X_clstr_ap = X_clstr[:,330:339]
X_clstr_to = X_clstr[:,339:341]
X_clstr_co = X_clstr[:,341:342]
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
mcc = make_scorer(matthews_corrcoef)
f1 = make_scorer(f1_score)

from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import PredefinedSplit
cv = PredefinedSplit(test_fold=splits)

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=100, n_jobs=8, random_state=42, class_weight="auto", max_depth=13)
```

```python
X_clstr_dict = {"ps": X_clstr_ps,
                "aa": X_clstr_aa,
                "do": X_clstr_do,
                "ss": X_clstr_ss,
                "ap": X_clstr_ap,
                "to": X_clstr_to,
                "co": X_clstr_co}
```

```python
%%time
from itertools import chain, combinations
def all_subsets(ss):
    return chain(*map(lambda x: combinations(ss, x), range(1, len(ss)+1)))

feature_tags = ["ps", "aa", "do", "ss", "ap", "to", "co"]
tags = []

clstr_features_scores = {}
for subset in all_subsets(feature_tags):
    tag = "_".join(list(subset))
    tags.append(tag)
    clstr_features_scores[tag] = {'mcc': [], 'f1': []}
    
    clstr_features = np.zeros((X_clstr_dict["ps"].shape[0], 1))
    for feat_set in subset:
        clstr_features = np.hstack((clstr_features, X_clstr_dict[feat_set]))  
    clstr_features = clstr_features[:,1:]
    
    for train, test in cv:
        rf_clf.fit(clstr_features[train], y_clstr[train])
        y_pred = rf_clf.predict(clstr_features[test])
        clstr_features_scores[tag]['mcc'].append(matthews_corrcoef(y_clstr[test], y_pred))
        clstr_features_scores[tag]['f1'].append(f1_score(y_clstr[test], y_pred))
        
        
    #clstr_features_scores[tag] = cross_val_score(rf_clf, clstr_features, y_clstr,
    #                                             scoring=mcc, cv=cv)

%store clstr_features_scores
%store tags
```

```python
import string
tags_list = []
mcc_scores = []
f1_scores = []
mcc_f1_scores = []
for tag in tags:
    tags_list.append(string.replace(tag, "_", " "))
    mcc_scores.append(np.array(clstr_features_scores[tag]['mcc']).mean())
    f1_scores.append(np.array(clstr_features_scores[tag]['f1']).mean())
    mcc_f1_scores.append((np.array(clstr_features_scores[tag]['mcc']).mean() + np.array(clstr_features_scores[tag]['f1']).mean()) / 2.)
    
    tmp_label = []
    for word in string.replace(tag, "_", " ").split():
        if word == "ps":
            tmp_label.append('1')
        elif word == "aa":
            tmp_label.append('2')
        elif word == "do":
            tmp_label.append('3')
        elif word == "ss":
            tmp_label.append('4')
        elif word == "ap":
            tmp_label.append('5')
        elif word == "to":
            tmp_label.append('6')
        elif word == "co":
            tmp_label.append('7')
    y_labels.append(", ".join(tmp_label))

best_indices_mcc = np.argsort(mcc_scores)
best_indices_f1 = np.argsort(f1_scores)
best_indices_mcc_f1 = np.argsort(mcc_f1_scores)

best_scores_mcc = []
best_labels_mcc = []
for index in reversed(best_indices_mcc):
    best_scores_mcc.append(mcc_scores[index])
    best_labels_mcc.append(y_labels[index])

best_scores_f1 = []
best_labels_f1 = []
for index in reversed(best_indices_f1):
    best_scores_f1.append(f1_scores[index])
    best_labels_f1.append(y_labels[index])

best_scores_mcc_f1 = []
best_labels_mcc_f1 = []
for index in reversed(best_indices_mcc_f1):
    best_scores_mcc_f1.append(mcc_f1_scores[index])
    best_labels_mcc_f1.append(y_labels[index])

%store best_scores_mcc
%store best_scores_f1
%store best_scores_mcc_f1
%store best_labels_mcc
%store best_labels_f1
%store best_labels_mcc_f1
```

```python
n_best = 20
plt.figure(figsize=(7, 4.33))
plt.barh(list(reversed(range(n_best))), best_scores_mcc[:n_best], height=0.7, align='center')
plt.yticks(list(reversed(range(n_best))), best_labels_mcc[:n_best])
plt.ylim(-0.8,19.8)
plt.xlabel("MCC score")
plt.ylabel("Feature groups")
plt.savefig('feature_selection_hbar_mcc.pdf', bbox_inches='tight')
plt.show()
```
