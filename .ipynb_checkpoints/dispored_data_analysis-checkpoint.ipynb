# Disopred Data Analysis

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Load data

For now, we will only use the **raw MorF data** to rule out any mistake I've could have made in the feature engineering.

The data was loaded from file and put into a numpy array in another notebook.

```python
X_morf = np.load("../data/X_disopred_morf.npy")
y_morf = np.load("../data/y_morf.npy")
print X_morf.shape, y_morf.shape
```

### Make classifiers, cross validation & scorer methods

We will use three classifiers: SGD, Linear SMV and SVM with an RBF kernel.

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
sgd_clf = SGDClassifier(class_weight="auto")
lin_svm_clf = LinearSVC(class_weight="auto")
rbf_svm_clf = SVC(kernel="rbf", class_weight="auto")
rf_clf = RandomForestClassifier(n_estimators=50,
                                random_state=42,
                                class_weight="auto")
```

We will also try some feature mapping again, with SGD and linear SVM.

The cross validation is stratified 5-fold

```python
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import StratifiedKFold
cv_morf = StratifiedKFold(y_morf, n_folds=5,
                     shuffle=True, random_state=42)
```

And MCC will be used as scorer matrix.

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)
```

## Learning Curve

```python
def plot_learning_curve(train_sizes, train_scores, test_scores,
                        title, ylim=None):

    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("MCC score")
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
%%time
from sklearn.learning_curve import learning_curve
sgd_train_sizes, sgd_train_scores, sgd_test_scores = learning_curve(
    sgd_clf, X_morf, y_morf, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_morf, scoring=mcc, n_jobs=5)
%store sgd_train_sizes
%store sgd_train_scores
%store sgd_test_scores
```

```python
%%time
from sklearn.learning_curve import learning_curve
svm_train_sizes, svm_train_scores, svm_test_scores = learning_curve(
    lin_svm_clf, X_morf, y_morf, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_morf, scoring=mcc, n_jobs=5)
%store svm_train_sizes
%store svm_train_scores
%store svm_test_scores
```

```python
%%time
from sklearn.learning_curve import learning_curve
rf_train_sizes, rf_train_scores, rf_test_scores = learning_curve(
    rf_clf, X_morf, y_morf, train_sizes=np.linspace(.01, 1.0, 10),
    cv=cv_morf, scoring=mcc, n_jobs=5)
%store rf_train_sizes
%store rf_train_scores
%store rf_test_scores
```

```python
title = "Learning Curves (Stochastic Gradient Descent)"
plot_learning_curve(sgd_train_sizes, sgd_train_scores, sgd_test_scores,
                    title, ylim=(0,1))
plt.show()
```

```python
title = "Learning Curves (Linear SMV)"
plot_learning_curve(svm_train_sizes, svm_train_scores, svm_test_scores,
                    title, ylim=(0,1))
plt.show()
```

```python
title = "Learning Curves (Random Forest)"
plot_learning_curve(rf_train_sizes, rf_train_scores, rf_test_scores,
                    title, ylim=(0,1))
plt.show()
```

We still have a *very* flat learning curve and the training and test score merge very quickly.

## Feature Mapping

But maybe we need to do some feature mapping to handle non-linear data.

### Without feature mapping

```python
%%time
sgd_morf_scores = cross_val_score(sgd_clf, X_morf, y_morf, cv=cv_morf,
                                  scoring=mcc, n_jobs=5)
%store sgd_scores
print "MCC: %.2f ± %.2f" % (sgd_scores.mean(), sgd_scores.std())
```

```python
%%time
lin_svm_morf_scores = cross_val_score(lin_svm_clf, X_morf, y_morf,
                                      cv=cv_morf, scoring=mcc, n_jobs=5)
%store lin_svm_morf_scores
print "MCC: %.2f ± %.2f" % (lin_svm_morf_scores.mean(), lin_svm_morf_scores.std())
```

```python
%%time
rf_morf_scores = cross_val_score(rf_clf, X_morf, y_morf,
                                 cv=cv_morf, scoring=mcc, n_jobs=5)
%store rf_morf_scores
print "MCC: %.2f ± %.2f" % (rf_morf_scores.mean(), rf_morf_scores.std())
```

### Now with feature mapping!

```python
from sklearn.pipeline import Pipeline
from sklearn.kernel_approximation import Nystroem
feature_map_nstrm = Nystroem()
nstrm_sgd = Pipeline([("feature_map", feature_map_nstrm),
                             ("SGD", sgd_clf)])
nstrm_svm = Pipeline([("feature_map", feature_map_nstrm),
                             ("svm", lin_svm_clf)])
nstrm_rf = Pipeline([("feature_map", feature_map_nstrm),
                            ("rf", rf_clf)])
```

```python
%%time
nstrm_sgd_scores = cross_val_score(nstrm_sgd, X_morf, y_morf,
                                   cv=cv_morf, n_jobs=5, scoring=mcc)
%store nstrm_sgd_scores
print "MCC: %.2f ± %.2f" % (nstrm_sgd_scores.mean(), nstrm_sgd_scores.std())
```

```python
%%time
nstrm_svm_scores = cross_val_score(nstrm_svm, X_morf, y_morf,
                                   cv=cv_morf, n_jobs=5, scoring=mcc)
%store nstrm_svm_scores
print "MCC: %.2f ± %.2f" % (nstrm_svm_scores.mean(), nstrm_svm_scores.std())
```

```python
%%time
nstrm_rf_scores = cross_val_score(nstrm_rf, X_morf, y_morf,
                                  cv=cv_morf, n_jobs=5, scoring=mcc)
%store nstrm_rf_scores
print "MCC: %.2f ± %.2f" % (nstrm_rf_scores.mean(), nstrm_rf_scores.std())
```

### Score summary

```python
from pandas import DataFrame
featmap_data = {
                "mcc_score": np.hstack((sgd_morf_scores, lin_svm_morf_scores,
                                        rf_morf_scores, nstrm_sgd_scores,
                                        nstrm_svm_scores, nstrm_rf_scores)),
                "classifier": (["SGD"] * 5 + ["Linear_SVM"] * 5
                               + ["Random Forest"] * 5) * 2,
                "feature_mapped": ["No"] * 15 + ["Yes"] * 15,
                "pca_transform": ["None"] * 30,
                "dataset": ["MoRF"] * 30
                }
featmap_data = DataFrame(featmap_data)
sns.factorplot("classifier", "mcc_score", "feature_mapped", featmap_data,
               kind="bar", palette="Paired")
```

Nope, feature mapping is not our friend.

## PCA

Let's apply some Principal Component Analasys, both to visualize the data and make it a bit faster to compute.

### Plot 2 most important dimensions

```python
from sklearn.decomposition import PCA
pca_2comp_morf = PCA(n_components=2)
pca_2comp_morf.fit(X_morf)
X_morf_2comp = pca_2comp_morf.transform(X_morf)
print "Retained dimensions: %i" % pca_2comp_morf.n_components_
print "Retained variance: %.2f" % sum(pca_2comp_morf.explained_variance_ratio_)
```

Let's split and plot the data!

```python
f, axes = plt.subplots(1, 3, figsize=(9, 3), sharex=True, sharey=True)
axes[0].scatter(X_morf_2comp[:,0], X_morf_2comp[:,1],
            c=y_morf, cmap=plt.cm.Paired, alpha=0.05)
axes[1].scatter(X_morf_2comp[y_morf==0,0], X_morf_2comp[y_morf==0,1],
            c=sns.xkcd_rgb["light grey blue"], alpha=0.05)
axes[2].scatter(X_morf_2comp[y_morf==1,0], X_morf_2comp[y_morf==1,1],
            c=sns.xkcd_rgb["auburn"], alpha=0.05)
f.tight_layout()
```

The distribution looks *very much alike* for both positive and negative examples...

### Retain 90% variance

Instead of reducing the data to two dimensions, let's **keep 90% of the variance**, and see how many dimensions we are left with.

```python
from sklearn.decomposition import PCA
pca_90_morf = PCA(0.90)
pca_90_morf.fit(X_morf)
X_morf_90 = pca_90_morf.transform(X_morf)
print "Retained dimensions: %i" % pca_90_morf.n_components_
print "Retained variance: %.2f" % sum(pca_90_morf.explained_variance_ratio_)
```
