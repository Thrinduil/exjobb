```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=10000, random_state=42, n_features=100, n_informative=20, n_classes=2)
print X.shape
print y.shape
```

```python
from sklearn.datasets import make_blobs
X_blobs, y_blobs = make_blobs(n_samples=1000, n_features=2, centers=4, cluster_std=1.0, random_state=42)
y_blobs[y_blobs==0] = -1
y_blobs[y_blobs==1] = 1
y_blobs[y_blobs==2] = -1
y_blobs[y_blobs==3] = 1
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=y_blobs, cmap=plt.cm.Paired)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=50, n_jobs=5, random_state=42, class_weight="auto")
rf_clf_oob = RandomForestClassifier(n_estimators=50, n_jobs=5, random_state=42, class_weight="auto", oob_score=True)
```

```python
from sklearn.cross_validation import StratifiedKFold
cv = StratifiedKFold(y, n_folds=5, shuffle=True, random_state=42)
```

```python
from sklearn.metrics import roc_curve, auc
from scipy import interp

def plot_roc_curve(X, y, cv):

    mean_tpr = 0.0
    mean_fpr = np.linspace(0, 1, 100)
    all_tpr = []
    
    for i, (train, test) in enumerate(cv):
        probas_ = rf_clf.fit(X[train], y[train]).predict_proba(X[test])
        fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
        mean_tpr += interp(mean_fpr, fpr, tpr)
        mean_tpr[0] = 0.0
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.3f)' % (i, roc_auc))
    
    plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
    
    mean_tpr /= len(cv)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    plt.plot(mean_fpr, mean_tpr, 'k--',
             label='Mean ROC (area = %0.3f)' % mean_auc, lw=2)
    
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    
    return plt
```

```python
plt = plot_roc_curve(X, y, cv)
plt.show()
```

# Tennis Data

```python
s = 1 #"sunny"
o = 2 #"overcast"
r = 3 #"rain"
h = 2 #"high"
n = 1 #"normal"
st = 2 #"strong"
we = 1 #"weak"
ye = 1 #"yes"
no = -1 #"no"
tennis_data = {"outlook": [s, s, o, r, r, r, o, s, s, r, s, o, o, r, r],
               "humidity": [h, h, h, h, n, n, n, h, n, n, n, h, n, h, n],
               "wind": [we, st, we, we, we, st, st, we, we, we, st, st, we, st, st],
               "play": [no, no, ye, ye, ye, no, ye, no, ye, ye, ye, ye, ye, no, no]}
tennis_data = pd.DataFrame(tennis_data)
```

```python
tennis_data
```

```python
from sklearn.tree import DecisionTreeClassifier
X = tennis_data.as_matrix(["humidity", "outlook", "wind"])
y = tennis_data.as_matrix(["play"])
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X, y)
```

```python
tree_clf.predict([3, 1, 2])
```

```python
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(tree_clf, out_file=dot_data)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_pdf("tennis.pdf")
```
