```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
```

```python
y_morf = np.load("../data/y_morf.npy")
X_myfeat_morf = np.load("../data/X_myfeat_morf.npy")
X_myfeat_morf_aa = X_myfeat_morf[:,300:320]
X_myfeat_morf_do = X_myfeat_morf[:,320:325]

y_pros = np.load("../data/y_pros.npy")
X_myfeat_pros = np.load("../data/X_myfeat_pros.npy")
X_myfeat_pros_aa = X_myfeat_pros[:,300:320]
X_myfeat_pros_do = X_myfeat_pros[:,320:325]

y_both = np.load("../data/y_both.npy")
X_myfeat_both = np.load("../data/X_myfeat_both.npy")
X_myfeat_both_aa = X_myfeat_both[:,300:320]
X_myfeat_both_do = X_myfeat_both[:,320:325]
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)

from sklearn.cross_validation import StratifiedKFold
cv_morf = StratifiedKFold(y_morf, n_folds=5,
                          shuffle=True, random_state=42)
cv_pros = StratifiedKFold(y_pros, n_folds=5,
                          shuffle=True, random_state=42)
cv_both = StratifiedKFold(y_both, n_folds=5,
                          shuffle=True, random_state=42)

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=50, random_state=42, class_weight="auto")
pl_rf_clf = make_pipeline(StandardScaler(), rf_clf)

from sklearn.cross_validation import cross_val_score
```

## aa_features

```python
%%time
morf_aa_scores = []
for morf_aa_feature in X_myfeat_morf_aa[:,0:].T:
    morf_aa_feature = morf_aa_feature.reshape((morf_aa_feature.shape[0],1))
    morf_aa_scores.append(cross_val_score(pl_rf_clf, morf_aa_feature, y_morf,
                                          scoring=mcc, cv=cv_morf, n_jobs=-1))
%store morf_aa_scores
```

```python
%%time
pros_aa_scores = []
for pros_aa_feature in X_myfeat_pros_aa[:,0:].T:
    pros_aa_feature = pros_aa_feature.reshape((pros_aa_feature.shape[0],1))
    pros_aa_scores.append(cross_val_score(pl_rf_clf, pros_aa_feature, y_pros,
                                          scoring=mcc, cv=cv_pros, n_jobs=-1))
%store pros_aa_scores
```

```python
%%time
both_aa_scores = []
for both_aa_feature in X_myfeat_both_aa[:,0:].T:
    both_aa_feature = both_aa_feature.reshape((both_aa_feature.shape[0],1))
    both_aa_scores.append(cross_val_score(pl_rf_clf, both_aa_feature, y_both,
                                          scoring=mcc, cv=cv_both, n_jobs=-1))
%store both_aa_scores
```

```python
aa_list = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
           'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
aa_data = {"aa": [], "mcc_score": [], "dataset": []}
aa_scores = [morf_aa_scores, pros_aa_scores, both_aa_scores]
datasets = ["MoRF", "ProS", "Both"]
for score_idx, aa_score in enumerate(aa_scores):
    for aa_idx, aa in enumerate(aa_list):
        for score in aa_score[aa_idx]:
            aa_data["aa"].append(aa)
            aa_data["mcc_score"].append(score)
            aa_data["dataset"].append(datasets[score_idx])
aa_data = DataFrame(aa_data)
```

```python
sns.factorplot("dataset", "mcc_score", "aa", data=aa_data, kind="bar", aspect=3, palette="Paired")
```

```python
%%time
pros_do_scores = []
for pros_do_feature in X_myfeat_pros_do[:,0:].T:
    pros_do_feature = pros_do_feature.reshape((pros_do_feature.shape[0],1))
    pros_do_scores.append(cross_val_score(pl_rf_clf, pros_do_feature, y_pros,
                                          scoring=mcc, cv=cv_pros, n_jobs=-1))
%store pros_do_scores
```

## Disorder automagic

```python
do_morf_dict = {"is_disorder": X_myfeat_morf_do[:,0].reshape((X_myfeat_morf_do.shape[0],1)),
                "score": X_myfeat_morf_do[:,1].reshape((X_myfeat_morf_do.shape[0],1)),
                "length": X_myfeat_morf_do[:,2].reshape((X_myfeat_morf_do.shape[0],1)),
                "start": X_myfeat_morf_do[:,3].reshape((X_myfeat_morf_do.shape[0],1)),
                "stop": X_myfeat_morf_do[:,4].reshape((X_myfeat_morf_do.shape[0],1))}

do_pros_dict = {"is_disorder": X_myfeat_pros_do[:,0].reshape((X_myfeat_pros_do.shape[0],1)),
                "score": X_myfeat_pros_do[:,1].reshape((X_myfeat_pros_do.shape[0],1)),
                "length": X_myfeat_pros_do[:,2].reshape((X_myfeat_pros_do.shape[0],1)),
                "start": X_myfeat_pros_do[:,3].reshape((X_myfeat_pros_do.shape[0],1)),
                "stop": X_myfeat_pros_do[:,4].reshape((X_myfeat_pros_do.shape[0],1))}

do_both_dict = {"is_disorder": X_myfeat_both_do[:,0].reshape((X_myfeat_both_do.shape[0],1)),
                "score": X_myfeat_both_do[:,1].reshape((X_myfeat_both_do.shape[0],1)),
                "length": X_myfeat_both_do[:,2].reshape((X_myfeat_both_do.shape[0],1)),
                "start": X_myfeat_both_do[:,3].reshape((X_myfeat_both_do.shape[0],1)),
                "stop": X_myfeat_both_do[:,4].reshape((X_myfeat_both_do.shape[0],1))}
```

```python
%%time
from itertools import chain, combinations
def all_subsets(ss):
  return chain(*map(lambda x: combinations(ss, x), range(1, len(ss)+1)))

feature_tags = ["is_disorder", "score", "length", "start", "stop"]
tags = []

morf_features_scores = {}
pros_features_scores = {}
both_features_scores = {}

for subset in all_subsets(feature_tags):
    tag = "_".join(list(subset))
    tags.append(tag)
    
    morf_features = np.zeros((do_morf_dict["is_disorder"].shape[0], 1))
    pros_features = np.zeros((do_pros_dict["is_disorder"].shape[0], 1))
    both_features = np.zeros((do_both_dict["is_disorder"].shape[0], 1))
    
    for feat_set in subset:
        morf_features = np.hstack((morf_features, do_morf_dict[feat_set]))
        pros_features = np.hstack((pros_features, do_pros_dict[feat_set]))
        both_features = np.hstack((both_features, do_both_dict[feat_set]))
        
    morf_features = morf_features[:,1:]
    pros_features = pros_features[:,1:]
    both_features = both_features[:,1:]
    
    morf_features_scores[tag] = cross_val_score(pl_rf_clf, morf_features, y_morf,
                                                scoring=mcc, cv=cv_morf, n_jobs=-1)
    pros_features_scores[tag] = cross_val_score(pl_rf_clf, pros_features, y_pros,
                                                scoring=mcc, cv=cv_pros, n_jobs=5)
    both_features_scores[tag] = cross_val_score(pl_rf_clf, both_features, y_both,
                                                scoring=mcc, cv=cv_both, n_jobs=3)
%store morf_features_scores
%store pros_features_scores
%store both_features_scores
%store tags
```

```python
feature_data = {"mcc_score": np.zeros((1,)),
                "dataset": [],
                "features": []
                }

for tag in tags:
    feature_data["mcc_score"] = np.hstack((feature_data["mcc_score"], morf_features_scores[tag]))
    feature_data["dataset"] = feature_data["dataset"] + ["MoRF"] * 5
    feature_data["features"] = feature_data["features"] + [tag] * 5
    
    feature_data["mcc_score"] = np.hstack((feature_data["mcc_score"], pros_features_scores[tag]))
    feature_data["dataset"] = feature_data["dataset"] + ["ProS"] * 5
    feature_data["features"] = feature_data["features"] + [tag] * 5
    
    feature_data["mcc_score"] = np.hstack((feature_data["mcc_score"], both_features_scores[tag]))
    feature_data["dataset"] = feature_data["dataset"] + ["Both"] * 5
    feature_data["features"] = feature_data["features"] + [tag] * 5
    
feature_data["mcc_score"] = feature_data["mcc_score"][1:]

feature_data = DataFrame(feature_data)
```

```python
sns.factorplot("dataset", "mcc_score", "features",
               feature_data, kind="bar", palette="Paired", aspect=6)
```

## ROC Curve

```python
from sklearn.metrics import roc_curve, auc
from scipy import interp

def plot_roc_curve(X, y, cv):

    mean_tpr = 0.0
    mean_fpr = np.linspace(0, 1, 100)
    all_tpr = []
    
    for i, (train, test) in enumerate(cv):
        probas_ = rf_clf.fit(X[train], y[train]).predict_proba(X[test])
        fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
        mean_tpr += interp(mean_fpr, fpr, tpr)
        mean_tpr[0] = 0.0
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.3f)' % (i, roc_auc))
    
    plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
    
    mean_tpr /= len(cv)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    plt.plot(mean_fpr, mean_tpr, 'k--',
             label='Mean ROC (area = %0.3f)' % mean_auc, lw=2)
    
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    
    return plt
```

```python
plt = plot_roc_curve(morf_aa_do_feat, y_morf, cv_morf)
plt.show()
```

```python
plt = plot_roc_curve(pros_aa_do_feat, y_pros, cv_pros)
plt.show()
```

```python
plt = plot_roc_curve(both_aa_do_feat, y_both, cv_both)
plt.show()
```

## Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
et_clf = ExtraTreesClassifier(n_estimators=50, random_state=42, class_weight="auto")
morf_et_scores = cross_val_score(et_clf, morf_aa_do_feat, y_morf,
                                 scoring=mcc, cv=cv_morf, n_jobs=5)
pros_et_scores = cross_val_score(et_clf, pros_aa_do_feat, y_pros,
                                 scoring=mcc, cv=cv_pros, n_jobs=5)
both_et_scores = cross_val_score(et_clf, both_aa_do_feat, y_both,
                                 scoring=mcc, cv=cv_morf, n_jobs=5)
```

```python
# Build a forest and compute the feature importances
et_clf.fit(morf_aa_do_feat, y_morf)
importances = et_clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in et_clf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(morf_aa_do_feat.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(morf_aa_do_feat.shape[1]), indices)
#plt.xlim([-1, 10])
plt.show()
```

```python
importances_array = np.zeros((1,1))
for tree in et_clf.estimators_:
    importances_vector = np.array(tree.feature_importances_).reshape((25,1))
    importances_array = np.vstack((importances_array, importances_vector))
importances_data = {"importance": list(importances_array[1:,:]),
                    "feature": feature_names * 50}
importances_data = DataFrame(importances_data)
```

```python
feature_names = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
                 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y',
                 "is_disorder", "score", "length", "start", "stop"]
indices = np.argsort(importances)[::-1]
sorted_feature_names = [feature_names[idx] for idx in indices]
sns.factorplot("feature", "importance", data=importances_data, kind="bar",
               aspect=2.5, x_order=sorted_feature_names)
```
