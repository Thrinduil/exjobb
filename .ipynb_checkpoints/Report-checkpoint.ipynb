## Stochastic Gradient Descent

Make pipeline that will **scale** the data before doing **Stochastic Gradient Descent**.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import make_pipeline
SGDclf = SGDClassifier(loss="hinge", class_weight='auto')
SGDclfPL = make_pipeline(StandardScaler(), SGDclf)
```

Run **Stochastic Gradient Descent** on the data.

```python
%%time
from sklearn.cross_validation import cross_val_score
sgd_scores_acc = cross_val_score(SGDclfPL, X, y, cv=cv,
                             scoring="accuracy", n_jobs=4)
%store sgd_scores_acc
print "Accuracy: %.2f +/- %.2f" % (sgd_scores_acc.mean(), sgd_scores_acc.std())
```

As only 1 in 50 examples are positive, accuracy is not a very good meadurment of success, because the model can reach 98% accuracy by classifying **all** examples as negative.

Instead we will use *Matthew's correlation coefficient* (MCC), a score that works for unbalanced data sets.

When using MCC +1 is perfect prediction, 0 is random and -1 is perfect inverse prediction.

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)
```

Let's just test the scoring method on some dummy data to see that it works.

```python
from sklearn.datasets import make_blobs
X_dummy, y_dummy = make_blobs(n_samples=100, n_features=2,
                              centers=2, cluster_std=4.0,
                              random_state=42)
plt.scatter(X_dummy[:,0], X_dummy[:,1], c=y_dummy, cmap=plt.cm.Paired)
plt.show()
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(SGDclfPL, X_dummy, y_dummy,
                         cv=5, scoring=mcc, n_jobs=4)
print "MCC: %.2f +/- %.2f" % (scores.mean(), scores.std())
```

So far so good.

## Validation curve

So maybe our bad results are due to a hyperparameter in the classifier. Let's see if we can find the optimal value for the regularization parameter $\alpha$:

```python
from sklearn.learning_curve import validation_curve

def plot_validation_curve(estimator, title, X, y, param_name, param_range,
                          ylim=None, cv=None, n_jobs=1):

    train_scores, test_scores = validation_curve(
    estimator, X, y, param_name=param_name, param_range=param_range,
    cv=cv, scoring=mcc, n_jobs=n_jobs)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.title(title)
    plt.xlabel(param_name)
    plt.ylabel("MCC score")
    if ylim != None:
        plt.ylim(ylim)
    plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
    plt.fill_between(param_range, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.2, color="r")
    plt.semilogx(param_range, test_scores_mean, label="Cross-validation score",
                 color="g")
    plt.fill_between(param_range, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.2, color="g")
    plt.legend(loc="best")
    
    return plt
```

```python
%%time
param_name = "alpha"
param_range = np.logspace(-10, 5, 8)
title = "Validation Curve with SGD"
plot_validation_curve(SGDclf, title, X, y, param_name, param_range,
                      cv=cv, n_jobs=5, ylim=(0, 1))
plt.show()
```

It is definitely possible to get a **worse** score by tuning the regularization parameter, but with the large margins of error, the default value of $\alpha = 10^{-4}$ seems to be about as good as it gets.

(**Note:** for some reasone I could not use my pipeline in the validation function, which means that the features are not scaled as in previous runs)

But it is a very strange looking validation curve. I would like it to look something like this:

```python
from IPython.display import Image
Image(url='http://scikit-learn.org/stable/_images/plot_validation_curve_001.png', width=500)
```

In the validation curve above, we see that for low $\gamma$ values both the training and CV score is low because the model is **too constrained** to fit the data properly. For high $\gamma$ values, the model is **too loose** so that the model overfits the training data and is not able to *generalize* to the test data, so the test score plummets.

But that is not what we see in the validation curve for my data. It seems it can't performe worse no matter how much we constrain the model, and for a loose model **both** the test and training score goes down.

But the good-looking validation curve was made with SVM and not SGD, so I ran an SGD on my first set of dummy data (the one with two distinct centers), and got this graph:

```python
param_name = "alpha"
param_range = np.logspace(-10, 5, 8)
title = "Validation Curve with SGD"
plot_validation_curve(SGDclf, title, X_dummy, y_dummy, param_name, param_range,
                      cv=5, n_jobs=5)
plt.show()
```

That looks like the validation curve for my data! (Except the score is much better...)
So maybe we should not be so upset about the strange-looking validation curve. We turn our attention to the learning curve instead.

## Learning Curve

```python
from sklearn import cross_validation
from sklearn.learning_curve import learning_curve


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.01, 1.0, 20)):

    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("MCC Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=mcc,
        n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
%%time
title = "Learning Curves (SGD)"
plot_learning_curve(SGDclfPL, title, X, y, cv=cv, n_jobs=5,
                    train_sizes=np.linspace(.01, 1.0, 10))
plt.show()
```

At the first data point, where we use only 1% of our training examples, we see that the Training score outperforms the CV score (although they are both very bad). As we go towards using 100% of our data, the training score falls sharply while the CV score rises very slowly until the two scores almost align.

But they align on a **very** bad score. Basicly no better than chance.

Maybe it's because our data is not linearly separable. Let's have a look at that.

## Non-linear data and kernel approximation

Let's create some non-linearly separable data:

```python
from sklearn.datasets import make_blobs
centers = np.array([[10, 10], [-10, -10], [10, -10], [-10, 10]])
X_blobs, y_blobs = make_blobs(n_samples=400, n_features=2,
                              centers=centers, cluster_std=4.0,
                              random_state=42)
y_blobs[y_blobs==0] = 1
y_blobs[y_blobs==1] = 1
y_blobs[y_blobs==2] = -1
y_blobs[y_blobs==3] = -1
```

```python
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=y_blobs, cmap=plt.cm.Paired)
plt.show()
```

First try to use a linear SVM:

```python
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
lin_svm = LinearSVC()
scores = cross_val_score(lin_svm, X_blobs, y_blobs,
                         cv=5, scoring=mcc)
print "MCC: %.2f ± %.2f" % (scores.mean(), scores.std())
```

As expected, the score is around 0. There just is no way to separate these points with a straight line, so around 50% will be miss-classified, which is no better than chance.

But an RBF kernel, which is a non-linear SVM should be able to do it.

```python
from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score
rbf_svm = SVC(kernel="rbf")
scores = cross_val_score(rbf_svm, X_blobs, y_blobs,
                         cv=5, scoring=mcc)
print "MCC: %.2f ± %.2f" % (scores.mean(), scores.std())
```

Yep. It can almost perfectly separate the points without any tweaking whatsoever.

But as my data set is so large, using an RBF kernel would take a good part of forever.

Instead, let's try kernel approximation with the Nystroem method!

```python
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.kernel_approximation import Nystroem
lin_svc = LinearSVC()
feature_map_nystroem = Nystroem()
nystroem_approx_svm = Pipeline([("feature_map", feature_map_nystroem),
                                        ("svm", lin_svc)])
scores = cross_val_score(nystroem_approx_svm, X_blobs, y_blobs,
                         cv=5, n_jobs=5, scoring=mcc)
print "MCC: %.2f ± %.2f" % (scores.mean(), scores.std())
```

With that feature mapping trick we got a pretty good score using a **linear** SVM.

Let's try it with SGD instead of SVM:

```python
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.kernel_approximation import Nystroem
feature_map_nystroem = Nystroem()
nystroem_approx_sgd = Pipeline([("feature_map", feature_map_nystroem),
                                        ("SGD", SGDclf)])
scores = cross_val_score(nystroem_approx_sgd, X_blobs, y_blobs,
                         cv=5, n_jobs=5, scoring=mcc)
print "MCC: %.2f ± %.2f" % (scores.mean(), scores.std())
```

Not excellent, but still better than nothing.

Let's try that with our own data!

```python
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline

feature_map_nystroem = Nystroem()
nystroem_approx_sgd = Pipeline([("feature_map", feature_map_nystroem),
                                ("sgd", SGDclf)])
sgd_scores_nstrm = cross_val_score(nystroem_approx_sgd, X, y, cv=cv,
                                   n_jobs=5, scoring=mcc)
%store sgd_scores_nstrm
print "MCC: %.2f ± %.2f" % (sgd_scores_nstrm.mean(), sgd_scores_nstrm.std())
```
