```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

X = cal_housing.data
y = cal_housing.target
feat_names = cal_housing.feature_names

X = X[:10,2].reshape(10,1)
y = y[:10]

x_plot = np.linspace(float(min(X)), float(max(X)), num=100)
X_plot = x_plot.reshape(100,1)

plt.scatter(X, y)

for degree in [1, 2, 3, 5, 10]:
    model = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=1))
    model.fit(X, y)
    y_plot = model.predict(X_plot)
    plt.plot(x_plot, y_plot, label="degree %d" % degree)
```

```python
from sklearn.linear_model import LinearRegression
plt.figure(figsize=(7, 4))
x_fake = np.linspace(1, 10, 5)
X_fake = np.linspace(1, 10, 100)
#y_fake = [1, 2, 3.5, 5.3, 7.3, 7.9, 7.0, 6.5, 6.2, 6]
y_fake = [1, 5.2, 10.5, 6.0, 1.3]
plt.scatter(x_fake, y_fake, label="Data")
for degree in [1, 2, 8]:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x_fake.reshape(5,1), y_fake)
    y_plot = model.predict(X_fake.reshape(100,1))
    plt.plot(X_fake, y_plot, label="Degree %d" % degree)
plt.legend(loc='best')
plt.savefig('bias_vs_variance_alt.pdf', bbox_inches='tight')
```

```python
sin_x = np.linspace(0, np.pi, num=10)
```

```python
sin_y = np.sin(sin_x)
```

```python
sin_y_noise = []
for y in sin_y:
    sin_y_noise.append(y + np.random.uniform(-0.15, 0.15))
```

```python
plt.scatter(sin_x, sin_y_noise)
plt.plot(sin_x, sin_y)
```

```python
from sklearn.linear_model import LinearRegression
plt.figure(figsize=(7, 4))
sin_x_long = np.linspace(0, np.pi, num=100)
plt.scatter(sin_x, sin_y_noise, label="Noisy data")
labels = ["High bias, low variance model", "Balanced model", "Low bias, high variance model"]
for idx, degree in enumerate([1, 2, 13]):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(sin_x.reshape(10,1), sin_y_noise)
    sin_y_long = model.predict(sin_x_long.reshape(100,1))
    plt.plot(sin_x_long, sin_y_long, label=labels[idx])
    #plt.plot(sin_x_long, np.sin(sin_x_long), label="True pattern")
plt.legend(loc='best')
plt.savefig('bias_vs_variance.pdf', bbox_inches='tight')
```

# My house data

```python
house_data = {"Selling price": [500, 1200, 1600, 1800, 1950],
              "House size":    [50, 100, 135, 175, 250],
              "Number of bedrooms": [1, 2, 2, 3, 4],
              "House colour": [1, 2, 1, 1, 2]}
house_data = pd.DataFrame(house_data)
```

```python
from sklearn.linear_model import LinearRegression
plt.figure(figsize=(7, 4.33))
x_long = np.linspace(0, 301, num=100)
plt.scatter(np.array(house_data["House size"]), np.array(house_data['Selling price']), label="House data", s=50)
labels = ["High bias, low variance model", "Balanced model", "Low bias, high variance model"]
for idx, degree in enumerate([1, 2, 7]):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(np.array(house_data["House size"]).reshape(5,1), np.array(house_data['Selling price']))
    y_long = model.predict(x_long.reshape(100,1))
    plt.plot(x_long, y_long, label=labels[idx])
plt.xlim((40,255))
plt.ylim((300, 2700))
plt.xlabel("House size (m$^2$)")
plt.ylabel("Selling price (1,000 SEK)")
plt.legend(loc='best')
plt.savefig('bias_vs_variance.pdf', bbox_inches='tight')
```
