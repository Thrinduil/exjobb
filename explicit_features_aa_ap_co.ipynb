```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set(context='paper', style='darkgrid', palette='deep', font='sans-serif', font_scale=1.5, rc=None)
```

```python
y_clstr = np.load("../data/y_clstr.npy")
X_clstr = np.load("../data/X_clstr.npy")
splits = np.load("../data/splits.npy")
X_clstr_aa = X_clstr[:,300:320]
X_clstr_ap = X_clstr[:,330:339]
X_clstr_co = X_clstr[:,341:342]
X_aa_ap_co = np.hstack((X_clstr_aa, X_clstr_ap, X_clstr_co))
```

```python
y_ref = np.load("../data/y_ref.npy")
X_ref = np.load("../data/X_ref.npy")
X_ref_aa = X_ref[:,300:320]
X_ref_ap = X_ref[:,330:339]
X_ref_co = X_ref[:,341:342]
X_ref_aa_ap_co = np.hstack((X_ref_aa, X_ref_ap, X_ref_co))
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import matthews_corrcoef
mcc = make_scorer(matthews_corrcoef)

from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import PredefinedSplit
cv = PredefinedSplit(test_fold=splits)

from sklearn.ensemble import ExtraTreesClassifier
et_clf = ExtraTreesClassifier(n_estimators=50, n_jobs=7, random_state=42, class_weight="auto")

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=50, n_jobs=7, random_state=42, class_weight="auto")

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

from sklearn.linear_model import SGDClassifier
sgd_clf = make_pipeline(StandardScaler(), SGDClassifier(n_jobs=7, random_state=42, class_weight="auto"))

from sklearn.svm import LinearSVC
svm_clf = make_pipeline(StandardScaler(), LinearSVC(class_weight="auto", random_state=42))
```

## Learning Curve

Unfortunatly we can't use the built-in learning curve function, because it will split the dataset in a way that don't respect the clustering. So we'll have to make our own learning curve.

```python
total_examples = len(X_aa_ap_co)
percent_of_dataset = np.linspace(0.01, 1, 10)
X_folds = {}
y_folds = {}
```

The plan is to first divide the data into 5 equally sized parts. This will not respect the clusters perfectly (as the proteins are of different sizes, the number of examples differ a bit between the folds), but I think it is good enough.

```python
for idx in range(5):
    X_folds[idx] = X_aa_ap_co[(total_examples / 5) * idx:(total_examples / 5) * (idx + 1), :]
    y_folds[idx] = y_clstr[(total_examples / 5) * idx:(total_examples / 5) * (idx + 1)]
```

This way we miss three examples at the very end, but I guess we can live with that for now.

Then, when we want to use a certain percentage of the whole dataset, we will build that from the 5 folds we made in the previous step. When the 5-fold cross-validation splits the data, it should split it right at the boundries we used when forming the set. If that makes any sense.

```python
%%time
from sklearn.cross_validation import KFold
test_scores = {}
train_scores = {}
for percent in percent_of_dataset:
    test_scores[percent] = []
    train_scores[percent] = []
    
    X_tmp = np.zeros((1, 30))
    y_tmp = np.zeros((1,))
    for idx in range(5):
        X_tmp = np.vstack((X_tmp, X_folds[idx][:len(X_folds[idx]) * percent, :]))
        y_tmp = np.hstack((y_tmp, y_folds[idx][:len(y_folds[idx]) * percent]))
    X_tmp = X_tmp[1:,:]
    y_tmp = y_tmp[1:]

    cv_tmp = KFold(len(y_tmp), n_folds=5, shuffle=False, random_state=42)
    for train, test in cv_tmp:
        et_clf.fit(X_tmp[train], y_tmp[train])
        test_scores[percent].append(mcc(et_clf, X_tmp[test], y_tmp[test]))
        train_scores[percent].append(mcc(et_clf, X_tmp[train], y_tmp[train]))
%store test_scores
%store train_scores
```

```python
test_scores_mean = []
test_scores_err = []
train_scores_mean = []
train_scores_err = []
for percent in percent_of_dataset:
    test_scores[percent] = np.array(test_scores[percent])
    train_scores[percent] = np.array(train_scores[percent])
    
    test_scores_mean.append(test_scores[percent].mean())
    test_scores_err.append(test_scores[percent].std())
    train_scores_mean.append(train_scores[percent].mean())
    train_scores_err.append(train_scores[percent].std())
```

```python
plt.errorbar(percent_of_dataset, test_scores_mean, test_scores_err, label="Test score")
plt.errorbar(percent_of_dataset, train_scores_mean, train_scores_err, label="Train score")
plt.legend()
```

```python
test_score_list = []
train_score_list = []
percent_list = []
for percent in percent_of_dataset:
    test_score_list.extend(test_scores[percent])
    train_score_list.extend(train_scores[percent])
    for i in range(5):
        percent_list.append(percent)
```

```python
data = {"MCC score": test_score_list + train_score_list,
        "Percent of dataset used": percent_list * 2,
        "Set": ["Test set"] * 50 + ["Training set"] * 50}
data = pd.DataFrame(data)
```

## Learning curve 2

```python
from sklearn.learning_curve import learning_curve

def plot_learning_curve(estimator, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 10)):
    
    plt.figure()
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("MCC score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
%%time
plot_learning_curve(et_clf, X_aa_ap_co, y_clstr, ylim=(0, 1), cv=cv)
plt.show()
```

## Evaluate with more metrics

```python
%%time
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import matthews_corrcoef
scores = {}
rounds = 5
for idx in range(rounds):
    scores[idx] = {}
    scores[idx]["mcc"] = []
    scores[idx]["prf1"] = []
    for train, test in cv:
        et_clf_rdm = ExtraTreesClassifier(n_estimators=50, n_jobs=7,
                                          class_weight="auto", random_state=idx,
                                          max_depth=35, min_samples_leaf=1)
        et_clf_rdm.fit(X_aa_ap_co[train], y_clstr[train])
        y_pred = et_clf_rdm.predict(X_aa_ap_co[test])
        scores[idx]["mcc"].append(matthews_corrcoef(y_clstr[test], y_pred))
        scores[idx]["prf1"].append(precision_recall_fscore_support(y_clstr[test], y_pred, average="binary"))
    print "Round %i done!" % idx
```

```python
mcc_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
for idx in range(rounds):
    mcc_scores.extend(scores[idx]["mcc"])
    for fold in range(5):
        precision_scores.append(scores[idx]["prf1"][fold][0])
        recall_scores.append(scores[idx]["prf1"][fold][1])
        f1_scores.append(scores[idx]["prf1"][fold][2])
mcc_scores = np.array(mcc_scores)
precision_scores = np.array(precision_scores)
recall_scores = np.array(recall_scores)
f1_scores = np.array(f1_scores)
print "Average MCC score: %.3f +- %.3f" % (mcc_scores.mean(), mcc_scores.std())
print "Average precision score: %.3f +- %.3f" % (precision_scores.mean(), precision_scores.std())
print "Average recall score: %.3f +- %.3f" % (recall_scores.mean(), recall_scores.std())
print "Average F1 score: %.3f +- %.3f" % (f1_scores.mean(), f1_scores.std())
```
